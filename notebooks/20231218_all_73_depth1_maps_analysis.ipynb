This notebook is to work on code snippets for analyzing the results of running my PS code on all 73 of the depth-1 maps that I had initally downloaded from Niagara. This is a very small subset of all the DR6 depth-1 maps, but it is a manageable number of maps to test things with.

The output code is angle_calc_1702693246.yaml. I want to do two things:
1. Plot all of the angles and errorbars as a timestream
2. Plot all spectra (in angle_calc_1702693246_spectra.npy) for EE and BB to see how similar they are across a range of depth-1 maps that have different sizes and depths.

```python
for i in range(len(yaml_dict['list_of_maps'])):
    map_name = yaml_dict['list_of_maps'][i]
    time = float(map_name.split('_')[1])
    #print(map_name)
    array = map_name.split('_')[2]
    angle = yaml_dict['angle_estimates'][i][0]
    errorbar = yaml_dict['angle_estimates'][i][1]
    if array=='pa4':
        plt.errorbar(time,angle,yerr=errorbar, fmt='bo', alpha=0.5)
    elif array=='pa5':
        plt.errorbar(time,angle,yerr=errorbar, fmt='ro', alpha=0.5)
    elif array=='pa6':
        plt.errorbar(time,angle,yerr=errorbar, fmt='go', alpha=0.5)
plt.legend(["PA4", "PA5", 'PA6'])
plt.title("All 73 Maps")
plt.ylabel('Angle [deg]')
plt.xlabel('Time [s since epoch]')
#plt.show()
plt.savefig(fig_output_path+"depth1_timestream_73_test_maps.png", dpi=300)
plt.close()
```

```python
start_index = 11
stop_index = 47
lmax = 2500

act_dr4_spectra_root = '/home/zbh5/act_analysis/scripts/dr4_tests/act_dr4_spectra/'
full_bins, full_centers = np.loadtxt(act_dr4_spectra_root+'BIN_ACTPOL_50_4_SC_low_ell', usecols=(0,2), unpack=True)
bins, centers = full_bins[:stop_index], full_centers[:stop_index-1]

# Load CAMB EE spectrum
camb_file = "/home/zbh5/act_analysis/scripts/camb_19413797_lensedtotcls.dat"
ell,DlEE,DlBB = np.loadtxt(camb_file, usecols=(0,2,3), unpack=True)
# Note that ell runs from 2 to 5400
ClEE = DlEE * 2 * np.pi / (ell*(ell+1.))
ClBB = DlBB * 2 * np.pi / (ell*(ell+1.))
max_ell = int(bins[-1])+2
digitized = np.digitize(ell[:max_ell], bins, right=True)
CAMB_ClEE_binned = np.bincount(digitized, ClEE[:max_ell].reshape(-1))[1:-1]/np.bincount(digitized)[1:-1]
CAMB_ClBB_binned = np.bincount(digitized, ClBB[:max_ell].reshape(-1))[1:-1]/np.bincount(digitized)[1:-1]
```

```python
ell_b = spectra_dict[maps[0]]['ell'][start_index:]

plt.plot(ell_b, CAMB_ClEE_binned[start_index:], 'r--')
for i in range(len(maps)):
    plt.semilogy(ell_b, spectra_dict[maps[i]]['E1xE1'][start_index:],alpha=0.3)
#    plt.semilogy(ell_b, spectra_dict[maps[i]]['E2xE2'][start_index:],alpha=0.3)
#    plt.semilogy(ell_b, spectra_dict[maps[i]]['E1xE2'][start_index:],alpha=0.3)
plt.ylabel("ClEE")
plt.xlabel("Ell")
plt.title("EE Autospectra for All 73 Depth-1 Maps")
#plt.title("EE Autospectra for Ref Map in the Footprint of All 73 Depth-1 Maps")
#plt.show()
plt.savefig(fig_output_path+"ee_autospectra_all73_test_maps.png", dpi=300)
#plt.savefig(fig_output_path+"ee_autospectra_all73_ref_maps.png", dpi=300)
```
